```python
import pandas as pd
import json

df = pd.read_csv('data/manor_data.csv',encoding='utf-8')
df
```

```python
# Create a list to store message objects
messages_list = []

# Iterate through the DataFrame and create message objects
for index, row in df.iterrows():
    system_message = {
        "role": "system",
        "content": "你是觀光工廠大呷麵本家元宇宙花園的人工智慧導覽員,負責簡短回答玩家的問題"
    }
    user_message = {
        "role": "user",
        "content": row['question']
    }
    assistant_message = {
        "role": "assistant",
        "content": row['answer']
    }
    # Create a message object with system, user, and assistant messages
    message_object = {
        "messages": [system_message, user_message, assistant_message]
    }
    # Append the message object to the list
    messages_list.append(message_object)

with open('messages2.json', 'w', encoding='utf-8') as json_file:
    # Write each message object as a separate JSON object in the list
    for message in messages_list:
        json.dump(message, json_file, ensure_ascii=False)
        json_file.write('\n')

print("Messages extracted and saved to messages.json")
```
